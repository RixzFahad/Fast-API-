# 🖥️ PuTTY – Quick Notes

## What is PuTTY?
- PuTTY is a free software tool
- Used to connect to remote computers/servers
- Works mainly using SSH (Secure Shell)

## Why is PuTTY used?

1. Remote Access
   - Connect to another system from your laptop
   - Control servers without being physically there

2. Server Management
   - Manage websites, cloud servers (AWS, VPS), Linux systems

3. File Transfer
   - PSCP → copy files
   - PSFTP → transfer files

4. Development Work
   - Run commands on remote machines
   - Used in backend, DevOps, networking

## Key Features
- Lightweight
- Easy to use
- Supports SSH, Telnet
- Secure connection

## Simple Example
You have a website on a server
→ Use PuTTY
→ Login using IP + password
→ Run commands to manage your site

## One-Line Summary
PuTTY = Tool to access and control remote servers securely

# 🔐 SSH Key – Quick Notes

## What is an SSH Key?
- SSH Key stands for Secure Shell Key
- It is an authentication method
- Used to connect to remote servers securely

## Why is it used?
- Login without using a password
- More secure than passwords
- Prevents unauthorized access

## How it works
- Two keys are created:
  1. Public Key → stored on server
  2. Private Key → kept on your computer
- Server checks the key instead of asking password

## Key Features
- High security
- Fast login
- No need to remember passwords

## One-Line Summary
SSH Key = Secure way to login to servers without password

# 🔑 .pem File – Quick Notes

## What is a .pem File?
- .pem stands for Privacy Enhanced Mail
- It is a private key file format
- Commonly used in Linux and servers

## Why is it used?
- Used for secure authentication
- Helps to access remote servers
- Works with SSH instead of passwords

## How it is used
- You download a .pem file from server/cloud (like AWS)
- Use it while connecting via SSH or PuTTY
- Acts as your identity to login

## Key Features
- Secure login
- No password required
- Used with SSH keys

## One-Line Summary
.pem = Private key file used to securely access servers

# 🔐 .ppk File – Quick Notes

## What is a .ppk File?
- .ppk stands for PuTTY Private Key
- It is a private key file used by PuTTY

## Why is it used?
- Used to securely connect to servers using PuTTY
- Alternative to password login
- Required when using PuTTY on Windows

## How it is used
- Convert .pem file → .ppk using PuTTYgen
- Load .ppk file in PuTTY while connecting
- Authenticate without password

## Key Features
- Used only in PuTTY
- Secure authentication
- Easy integration with SSH

## One-Line Summary
.ppk = PuTTY private key file used for secure server login
