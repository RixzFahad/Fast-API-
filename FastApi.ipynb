**FastAPI Methods**

*FastAPI methods define what action your API will perform when a client sends a request.*

---

**1. GET**

- Used to **read or fetch data**
- It is used when you want to show data
- It should not normally modify data

**Example:** `@app.get("/users")`  
*Meaning:* Fetch all users

---

**2. POST**

- Used to **create new data**
- It sends data from client to server

**Example:** `@app.post("/users")`  
*Meaning:* Create a new user

---

**3. PUT**

- Used to **fully update existing data**
- It usually replaces the old data completely

**Example:** `@app.put("/users/{id}")`  
*Meaning:* Update full user data by id

---

**4. PATCH**

- Used to **partially update data**
- It changes only specific fields

**Example:** `@app.patch("/users/{id}")`  
*Meaning:* Update only selected user fields

---

**5. DELETE**

- Used to **remove data**
- It deletes a specific record

**Example:** `@app.delete("/users/{id}")`  
*Meaning:* Delete user by id

---

**6. OPTIONS**

- Used to check **which methods are allowed**
- Mostly used for API communication details

**Example:** `@app.options("/users")`  
*Meaning:* Show allowed methods for `/users`

---

**7. HEAD**

- Similar to GET
- Returns only **headers**
- Does not return response body

**Example:** `@app.head("/users")`  
*Meaning:* Check headers of `/users`

---

**8. TRACE**

- Used for **debugging**
- Shows how request reaches the server
- Rarely used in real projects

**Example:** `@app.trace("/users")`  
*Meaning:* Trace request path

---

**9. WebSocket**

- Used for **real-time communication**
- Not a normal HTTP method
- Used in chat apps, notifications, live updates

**Example:** `@app.websocket("/chat")`  
*Meaning:* Open live connection on `/chat`

---

**10. Quick Summary**

- `GET` → Read data
- `POST` → Create data
- `PUT` → Full update
- `PATCH` → Partial update
- `DELETE` → Delete data
- `OPTIONS` → Allowed methods
- `HEAD` → Headers only
- `TRACE` → Debugging
- `WebSocket` → Real-time connection

---

**11. Most Used in Projects**

- `GET`
- `POST`
- `PUT`
- `PATCH`
- `DELETE`

*These are the main CRUD methods used in real FastAPI projects.*

**Uvicorn**

*Uvicorn is a fast ASGI server used to run FastAPI applications.*

---

**What it does**

- It starts your FastAPI app
- It handles incoming requests from the browser or API client
- It sends responses back to the user
- It works like the engine that runs your backend app

---

**Why it is used**

- FastAPI itself is a framework
- To actually run that framework, you need a server
- Uvicorn is that server

---

**Simple understanding**

- `FastAPI` = builds the API
- `Uvicorn` = runs the API

*Without Uvicorn or another server, your FastAPI app will not run properly.*

---

```bash
uvicorn main:app --reload

---

**1. Path Parameters**

- Used to send values inside the URL
- Helpful when you want data for a specific item

**Example:** `@app.get("/users/{id}")`  
*Meaning:* get user by id

---

**2. Query Parameters**

- Used to send values after `?` in the URL
- Helpful for search, filter, sorting, pagination

**Example:** `@app.get("/users?active=true")`  
*Meaning:* get active users

---

**3. Request Body**

- Used when client sends data to the API
- Mostly used with POST, PUT, PATCH
- FastAPI reads JSON body very easily using models

**Example:** `def create_user(user: User)`  
*Meaning:* receive user data from request body

---

**4. Pydantic Models**

- Used for data validation
- Helps define input structure clearly
- Makes sure data types are correct

**Example:**  
`class User(BaseModel):`  
`    name: str`  
`    age: int`

*Meaning:* user must send name as text and age as number

---

**5. Response Model**

- Used to control what data your API returns
- Helps with validation, filtering, and documentation
- Very important for clean APIs

**Example:** `@app.get("/users", response_model=list[UserOut])` :contentReference[oaicite:1]{index=1}  
*Meaning:* return only output fields defined in `UserOut`

---

**6. Status Codes**

- APIs should return proper status codes
- These tell client what happened

**Common ones:**

- `200` → success
- `201` → created
- `400` → bad request
- `404` → not found
- `500` → server error

**Example:** `@app.post("/users", status_code=201)`  
*Meaning:* user created successfully

---

**7. Dependency Injection**

- One of FastAPI’s most important features
- Used to share logic like database connection, authentication, validation
- Makes code reusable and clean

**Example:** `user: User = Depends(get_current_user)`  
*Meaning:* get current user from shared dependency

*FastAPI has a dedicated dependency system in its tutorial structure and uses it widely across security and shared logic patterns.* :contentReference[oaicite:2]{index=2}

---

**8. APIRouter**

- Used to split routes into multiple files
- Very useful in bigger projects
- Similar idea to organizing users, products, orders separately

**Example:** `router = APIRouter()`  
*Meaning:* create grouped routes for one module

*FastAPI recommends structuring bigger applications into multiple files and route groups instead of keeping everything in one file.* :contentReference[oaicite:3]{index=3}

---

**9. Middleware**

- Runs before request reaches route and before response goes back
- Useful for logging, timing, auth checks, custom headers

**Example:** `@app.middleware("http")`  
*Meaning:* run logic for every request and response

*FastAPI defines middleware as logic that works with every request before path operations and with every response before returning it.* :contentReference[oaicite:4]{index=4}

---

**10. CORS**

- Needed when frontend and backend run on different ports or domains
- Very common in React + FastAPI projects
- Without CORS setup, browser may block requests

**Example idea:** allow frontend like `http://localhost:3000`

*FastAPI includes built-in support for CORS configuration in its docs and middleware-based setup.* :contentReference[oaicite:5]{index=5}

---

**11. Background Tasks**

- Used when work should happen after response is sent
- Helpful for email sending, logging, notifications, file processing

**Example:** `background_tasks.add_task(send_email, email)`  
*Meaning:* send email in background after response

*FastAPI documents background tasks specifically for work that should happen after returning a response, so the client does not need to wait for it.* :contentReference[oaicite:6]{index=6}
